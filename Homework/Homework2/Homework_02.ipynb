````python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

        def __str__(self):
            return f"{self.name} is {self.age} years old."
        
class pet(Animal):
    def __init__(self, name, age, is_friendly):
        super().__init__(name, age)
        self.is_friendly = is_friendly

        def __str__(self):
            return f"{self.name} is {self.age} years old. Cuddly: {self.cuddly}."
        
class farm_Animal(Animal):
    def __init__(self, name, age, for_meat, for_leather, helps_farming):
        super().__init__(name, age)
        self.for_meat = for_meat
        self.for_leather = for_leather
        self.helps_farming = helps_farming

        def __str__(self):
            return f"{self.name} is {self.age} years old. For meat: {self.for_meat}, for leather: {self.for_leather}, helps farming: {self.helps_farming}."
        
class Cow(farm_Animal):
    def __init__(self, name, age):
        super().__init__(name, age, for_meat=True, for_leather=True, helps_farming=True)

class Pig(farm_Animal):
    def __init__(self, name, age):
        super().__init__(name, age, for_meat=True, for_leather=False, helps_farming=False)

class mini_pig(pet):
    def __init__(self, name, age):
        super().__init__(name, age, is_friendly=True)


farm = []

def add_animal():
    print("\nChoose an animal to add: (cow, pig, mini_pig)")
    animal = input(">")
    name = input("Name: ")
    age = input("Age: ")
    if animal == "cow":
        farm.append(Cow(name, age))
    elif animal == "pig":
        farm.append(Pig(name, age))
    elif animal == "mini_pig":
        farm.append(mini_pig(name, age))
    else:
        print("Invalid animal.")
    print("Animal added.")

def remove_animal():
    print("\nChoose an animal to remove: (cow, pig, mini_pig)")
    animal = input(">")
    name = input("Name: ")
    for i in farm:
        if i.name == name:
            farm.remove(i)
            print("Animal killed.")
            return
        print("Animal not found.")

        index = int(input("Enter the number of the animal to remove: "))
        if 0 <= index < len(farm):
            print (f"Removing {farm[index].name}")
        else:
            print("Invalid index.")
    
while True:
    print("\n1. Add an animal\n2. Remove an animal\n3. View all animals\n4. Exit")
    choice = input("Choose an option: ")

    if choice == "1":
        add_animal()
    elif choice == "2":
        remove_animal()
    elif choice == "3":
        if farm:
            for animal in farm:
                print(animal)
        else:
            print("No animals on the farm.")
    elif choice == "4":
        break
    else:
        print("Invalid choice. Try again!")
````
